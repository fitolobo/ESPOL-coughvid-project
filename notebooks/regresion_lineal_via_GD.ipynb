# Regresión Lineal via Gradiente Descendente

- $C$: costo
- $l$: loss (pérdida) --> $\mathcal{L}$
- $D$: $\{(x_i,y_i)\}_{1}^{n}$ --> $x_i$ es un vector que contiene características (features) e $y_i$: valor objetivo (número real) en el caso de problemas de regresión. Y cuando este valor $y_i$ es un entero lo llamamos de problema de clasificación y generalmente se resuelve via regresión logística. 
- $h$: es el modelo paramétrico ($\theta$) a veces theta se denomina con la letra $w$ por la palabra en inglés *weights*

- Intuitivamente el problema tiene por objetivo: $y_i \approx h_{\theta}(x_i)$

- $l$ o el loss va a medir el error cometido por nuestro modelo para un par del conjunto $D$. 
- $C$ mide el error cometido por el modelo para la totalidad de los datos (para un conjunto/vector de parametros $\theta$ fijo)


$C(D,\theta) = \displaystyle \sum_{i=1}^{n} l(y_i,h_{\theta}(x_i))$

En la literatura dependiendo de la naturaleza del problema existen diferentes $l$. 
- $l(y_i,h_{\theta}(x_i)) = \left(y_i - h_{\theta}(x_i)\right)^{2}$ 

- Necesitamos derivar esta expresión pues queremos conocer los valores de $\theta$ para los cuales alcanzamos un mínimo en la suma de errores cometidos por nuestro modelo. 

## Gradiente Descendente 

$\theta^{k+1} = \theta^{k} -\alpha \nabla_{\theta} C(D,\theta)$ 

Calculando el gradiente: 

$\nabla_{\theta} C(D,\theta) = \nabla_{\theta}  \displaystyle \sum_{i=1}^{n} l(y_i,h_{\theta}(x_i)) = \nabla_{\theta} \displaystyle \sum_{i=1}^{n} \left(y_i - h_{\theta}(x_i)\right)^{2}$ 

- $h_{\theta}(x_i) = \theta_0 + \theta_1 x_i$
- $h_{\theta}(x_i) = \theta_0 + \theta_1 x_i + \theta_2 x_i^{2}$
- $h_{\theta}(x_i) = \mathcal{M}(x_i)$ el mapa $\mathcal{M}$ es una función compuesta y en general esa composición es *profunda*. 
- $h_{\theta}(x_i) = \sigma\left(\sigma\left(\dots \sigma\left(A x_i + b\right)\right)\right)$ --> objetivo del próximo video


$\nabla_{\theta} C(D,\theta) = \nabla_{\theta}  \displaystyle \sum_{i=1}^{n} l(y_i,h_{\theta}(x_i)) = \nabla_{\theta} \displaystyle \sum_{i=1}^{n} \left(y_i - \theta_0 - \theta_1 x_i\right)^{2}$ 

- A veces se define la función de costo con una fracción multiplicando de forma escalar por conveniencia y se utiliza $\frac{1}{2}$

$\frac{1}{2}\nabla_{\theta} C(D,\theta) = \frac{1}{2}\nabla_{\theta}  \displaystyle \sum_{i=1}^{n} l(y_i,h_{\theta}(x_i)) = \frac{1}{2}\nabla_{\theta} \displaystyle \sum_{i=1}^{n} \left(y_i - \theta_0 - \theta_1 x_i\right)^{2}$ 

- Derivando: 

1 - Respecto a $\theta_0$: $\cancel{\frac{2}{2}} \displaystyle \sum_{i=1}^{n} \left(y_i - \theta_0 - \theta_1 x_i\right) \cdot \left(-1\right) = - \displaystyle \sum_{i=1}^{n} \left(y_i - \theta_0 - \theta_1 x_i\right)$

2 - Respecto a $\theta_1$: $- \displaystyle \sum_{i=1}^{n} \left(y_i - \theta_0 - \theta_1 x_i\right)\cdot x_i$


# Reemplazando en la ecuación de GD 

$\theta^{k+1} = \theta^{k} -\alpha \nabla_{\theta} C(D,\theta)$ 

- Forma vectorial 

$\begin{bmatrix} \theta_0 \\ \theta_1\end{bmatrix}^{k+1} = \begin{bmatrix} \theta_0 \\ \theta_1 \end{bmatrix}^{k} - \alpha \cdot \begin{bmatrix} - \displaystyle \sum_{i=1}^{n} \left(y_i - \theta_0 - \theta_1 x_i\right) \\ - \displaystyle \sum_{i=1}^{n} \left(y_i - \theta_0 - \theta_1 x_i\right)\cdot x_i \end{bmatrix}^{k}$

- EL supra indice $k$ es porque los valores de $\theta$ en el gradiente son los correspondientes a la iteración actual. 
- $\alpha$ es un parámetro que tu debes escoger (hiperparámetro).

### Versión sin módulos (solo un método)

````python
from typing import Union 

def metodo_gd(x: np.array ,y: np.array, theta: np.array, alpha: float, num_iteraciones: int) -> Union[np.array,np.array]:
    ''' 
    inputs:
    x: es el vector de caracteristicas (numpy array) de dimension n 
    y: es el vector de numeros reales objetivo de la regresion
    alpha: 
    num_iteraciones: 
    outputs:
    theta: es el vector de parametros estimados por el gradiente descendente 
    '''

    # info que podemos guardar 
    historia_del_error = []

    # En la tarea: 
    #epsilon = 0.00001 
    # while (error > epsilon) or (num_max_iteraciones < 100): 

    # aqui hay opciones: la persona determina cuantas veces repetimos esto (ingenua)
    # nosotros intentamos establecer un epsilon para detener la iteración --> el epsilon tiene que tener relacion con el error

    # Cuando iteramos utilizamos TODOS los datos (GD)
    # Pero! podriamos utilizar un dato por vez para refrescar los thetas (SGD)
    for i in range(num_iteraciones):
        print(f"Iteración número: {i+1}")

        # calculo el error
        error = y - (theta[0]+theta[1]*x)
        
        # calculo los gradientes
        grad_0 = -1*np.sum(error)
        grad_1 = -1*np.sum(error*x)
        
        # gradiente descendente (tarea es escribir esta iteracion pero vectorialmente)
        theta[0] = theta[0] - alpha * grad_0
        theta[1] = theta[1] - alpha * grad_1
        
        historia_del_error.append(np.sum(error**2))
    
    return  theta, historia_del_error
````

## Version Estocástica 

````python
from typing import Union 

def metodo_sgd(x: np.array ,y: np.array, theta: np.array, alpha: float, epoch: int) -> Union[np.array,np.array]:
    ''' 
    inputs:
        x: es el vector de caracteristicas (numpy array) de dimension n 
        y: es el vector de numeros reales objetivo de la regresion
        
    outputs:
        theta: es el vector de parametros estimados por el gradiente descendente +
    '''

    # info que podemos guardar 
    error_por_ejemplo = []
    historia_del_error = []

    # midiendo la cantidad de ejemplos 
    n = len(y)
    
    # En la tarea: 
    #epsilon = 0.00001 
    # while (error > epsilon) or (num_max_iteraciones < 100): 
    
    # aqui hay opciones: la persona determina cuantas veces repetimos esto (ingenua)
    # nosotros intentamos establecer un epsilon para detener la iteración --> el epsilon tiene que tener relacion con el error
    
    # Cuando iteramos utilizamos TODOS los datos (GD)
    # Pero! podriamos utilizar un dato por vez para refrescar los thetas (SGD)
    for i in range(epoch):
        print(f"Epoch número: {i+1}")
        for j in range(n):
            # calculo el error
            error = y[j] - (theta[0]+theta[1]*x[j])
            
            # calculo los gradientes
            grad_0 = -1*error
            grad_1 = -1*error*x[j]
            
            # gradiente descendente 
            theta[0] = theta[0] - alpha * grad_0
            theta[1] = theta[1] - alpha * grad_1
            
            error_por_ejemplo.append(error**2)
        historia_del_error.append(np.sum(error_por_ejemplo))
    
    return  theta, historia_del_error, error_por_ejemplo
````

````python
theta = np.random.rand(2)

# Hiperparametro 
alpha = 0.00000001 #OBS: el problema es la sensibilidad del hiperparametro 
epoch = 10000

theta, historia_del_error, error_por_ejemplo = metodo_sgd(x,y, theta, alpha, epoch)
````
